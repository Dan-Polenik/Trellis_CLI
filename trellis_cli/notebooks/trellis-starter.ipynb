# Trellis starter
Flink at `jobmanager:8081`. Pulsar at `localhost:8080` from host. Inside network use `host.containers.internal:8080` on mac or `localhost` on linux.

```scala
import $ivy.`org.apache.beam:beam-runners-flink-1:2.57.0`
import $ivy.`com.spotify::scio-core:0.14.9`
println("Scala kernel up. You can now add your Beam args with FlinkRunner.")
val flinkArgs = Array("--runner=FlinkRunner", "--flinkMaster=jobmanager:8081")
```

```scala
import $ivy.`com.softwaremill.sttp.client3::core:3.9.6`
import sttp.client3._
val pulsarUrl = sys.props.getOrElse("pulsar.http", "http://host.containers.internal:8080")
val backend = HttpURLConnectionBackend()
val r = basicRequest.get(uri"$pulsarUrl/admin/v2/brokers/health").send(backend)
println(s"Pulsar HTTP health status: ${r.code}")
```
